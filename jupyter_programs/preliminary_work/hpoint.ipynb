```python
import nltk
import os
import sys
from nltk.corpus.reader.plaintext import PlaintextCorpusReader
from nltk.probability import FreqDist
from nltk.stem import WordNetLemmatizer

def hpoint(text):


    lemmatizer = WordNetLemmatizer()

    #porter = nltk.PorterStemmer()
    list_v = ([lemmatizer.lemmatize(t) for t in text])

    fdist = FreqDist(list_v)
    new_list = fdist.most_common(len(fdist))

    count = 0


    for words in new_list:
        if words[0] == ',':
            new_list.remove(words)

    for words in new_list:
        if words[0] == '.':
            new_list.remove(words)

    for words in new_list:
        if words[0] == '"':
            new_list.remove(words)

    for words in new_list:
        if words[0] == '(':
            new_list.remove(words)

    for words in new_list:
        if words[0] == ')':
            new_list.remove(words)

    for words in new_list:
        if words[0] == '-':
            new_list.remove(words)

    for words in new_list:
        if words[0] == '--':
            new_list.remove(words)

    for words in new_list:
        if words[0] == ':':
            new_list.remove(words)


    for words in new_list:
        if words[0] == 'SYM':
            new_list.remove(words)

    for words in new_list:
        if words[0] == '``':
            new_list.remove(words)


    for words in new_list:
        if words[0] == 's':
            new_list.remove(words)
    

    for words in new_list:
        if words[0] == "'":
            new_list.remove(words)

    for words in new_list:
        if words[0] == ';':
            new_list.remove(words)


    for words, nextw in zip(new_list, new_list[1:]):

        count += 1
        
        if count == words[1]:
            print('hpoint '),
            print(count)
            print(new_list[:count])
            break;

        if (count < words[1]) and (count+1 > nextw[1]):
            print('hpoint alternative ' ),
            alt = (((words[1]) * (count+1)) - ((nextw[1]) * count))/ (((count+1)-count) + ((words[1])-(nextw[1])))
            print (alt)
            print(new_list[:alt])
            break;


def main():

    nltk.download('wordnet')

    file_name = sys.argv[1]
    newcorpus = PlaintextCorpusReader('data', '.*')
    files = newcorpus.fileids()

    for f in files:
        print(f)
        hpoint(newcorpus.words(f))
        
if __name__ == "__main__":
    main()





```
